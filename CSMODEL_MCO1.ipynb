### **I. Dataset Description**

> **Data Collection Process and Provenance**

- The **"Student Performance Factors"** dataset is hosted on Kaggle and was uploaded by the user **"Lai Ng."** A thorough investigation into the original source or methodology reveals no explicit, publicly documented provenance, such as a formal academic research paper, a detailed survey instrument, a specified sampling frame, or any institutional review board approval.
- Given the datasetâ€™s structure, specifically the clean, uniform distribution of many categorical variables (e.g., the balanced counts for Low, Medium, and High categories across different student attributes), it is highly probable that this dataset is synthetic. It was likely generated artificially using a controlled algorithm to simulate a real-world scenario for the purpose of teaching, learning, and practicing machine learning and statistical modeling techniques.

---



> **Implications of the Data Collection Method on Conclusions and Insights**
- The lack of verifiable provenance has several profound methodological and analytical implications that must be explicitly acknowledged in any generated conclusions or insights:
1. **Limited External Validity (Generalizability):**
- If the data is synthetic, the relationships observed between
variables (e.g., the predictive strength of Hours_Studied on Exam_Score) are ultimately **artifacts of the generating algorithm**, not necessarily reflections of actual, complex human behavior in a diverse student population.
- Any conclusions (such as specific policy recommendations or the ranking of influential factors) **cannot be reliably generalized** or applied to real-world educational systems, different schools, or students outside of the dataset's hypothetical constraints. The research should therefore be framed as a *proof of concept* or a study to validate an analytical technique using a simulated environment.

2. **Uncertainty Regarding Measurement Bias and Data Integrity:**
- For quantitative variables (Hours_Studied, Attendance), it is unknown whether the data was collected via student self-reporting (prone to social desirability bias) or objective tracking. Similarly, for categorical factors, the definition of the levels (e.g., the threshold for "High" Parental_Involvement) is not standardized. This ambiguity prevents a critical assessment of measurement error, and it means the variables may not accurately reflect the true underlying concepts they intend to capture.

3. **Inability to Infer Causality (Association vs. Causation):**
- Since this dataset is an observational snapshot, and not the result of a controlled, randomized experiment, it is fundamentally impossible to assert that a change in an input factor causes a change in the Exam_Score. Conclusions drawn from the model must strictly be limited to discussing predictive associations, correlations, and explanatory power. For instance, one can state that Tutoring_Sessions is a strong predictor of success, but not that attending more sessions causes a higher score.

4. **Ethical Responsibility in Reporting:**
- Utilizing data of unknown or synthetic origin necessitates absolute transparency in the final paper. Failing to clearly state these limitations risks misleading readers into believing that the results are grounded in empirical evidence from a real student body. This could potentially lead to the misallocation of educational resources or the implementation of flawed intervention policies based on simulated, rather than validated, data.

### **Discussion of Variables**
*The dataset contains 20 variables, all of which are described below:*

- **Hours_Studied** - A numerical variable representing the average number of hours the student spends actively studying per week.
- **Attendance** - A numerical variable representing the percentage of classes the student has attended.
- **Parental_Involvement** - A categorical variable indicating the perceived level of parental engagement in the student's education, categorized as: Low, Medium, or High.
- **Access_to_Resources** - A categorical variable describing the student's availability of and access to educational resources, such as books, computers, and learning materials, categorized as: Low, Medium, or High.
- **Extracurricular_Activities** - A binary (Yes/No) variable indicating whether the student participates in organized activities outside of regular classroom instruction.
- **Sleep_Hours** - A numerical variable indicating the average number of hours of sleep the student gets per night.
- **Previous_Scores** - A numerical variable representing the scores obtained by the student in previous, presumably standardized, exams or assessments.
- **Motivation_Level** - A categorical variable reflecting the student's self-reported or observed level of internal drive for academic success, categorized as: Low, Medium, or High.
- **Internet_Access** - A binary (Yes/No) variable indicating the availability of reliable internet access for study purposes in the student's home environment.
- **Tutoring_Sessions** - A numerical variable representing the number of formal tutoring sessions the student has attended per month.
- **Family_Income** - A categorical variable reflecting the socio-economic status of the student's family, categorized as: Low, Medium, or High.
- **Teacher_Quality** - A categorical variable reflecting the perceived quality or effectiveness of the teachers the student interacts with, categorized as: Low, Medium, or High.
- **School_Type** - A categorical variable classifying the type of school the student attends: Public or Private.
- **Peer_Influence** - A categorical variable indicating the perceived effect of the student's peers on their academic focus and performance, categorized as: Positive, Neutral, or Negative.
- **Physical_Activity** - A numerical variable representing the average number of hours the student engages in physical exercise or activity per week.
- **Learning_Disabilities** - A binary (Yes/No) variable indicating the presence of formally recognized learning disabilities or special educational needs.
- **Parental_Education_Level** - A categorical variable representing the highest level of education achieved by the student's parents or guardians, categorized as: High School, College, or Postgraduate.
- **Distance_from_Home** - A categorical variable describing the physical distance or commute time from the student's home to the school, categorized as: Near, Moderate, or Far.
- **Gender** - A categorical variable indicating the student's gender, categorized as: Male or Female.
- **Exam_Score** - The target variable for analysis; a numerical score representing the student's final performance on the main examination.

### **III. Exploratory Data Analysis Questions**

In our exploratory data analysis, the main goal is to determine which factors could correlate to the academic performance of the students. With this, we tried determining the general performance through central tendency scores and the correlation of socio-economic factors which are the students' access to internet connection and their family income, their study behaviors which are the amount of hours they studied and their access to learning materials to their exam scores. Overall, we find that there is a correlation between the study behaviors of the students to their exam scores or overall academic performance.

**1. What is the general performance of the students that is represented through the average and the spread?**

We will be geting the central tendencies of the scores mean, mode, and median. We will also be getting the spread by looking at standard deviation and the minimum and maximum scores

```python
scores = student_df['Exam_Score']

mean = scores.mean()
mode = scores.mode()[0]
median = scores.median()
std = scores.std()
min = scores.min()
max = scores.max()

#summary data to make it into a table format
summary_data = {
    'Measure':[
        'Mean',
        'Mode',
        'Median',
        'Standard Deviation',
        'Minimum',
        'Maximum'
    ],
    'Scores':[
        round(mean, 2),
        round(mode, 2),
        round(median, 2),
        round(std, 2),
        round(min, 2),
        round(max, 2)
    ]
}

summary_table = pd.DataFrame(summary_data)
print('General Performance of Students through Exam Scores')
display(summary_table)
```

The data has been normalized especially since the mean, mode, and median are almost equal to one another

**Data Visualization** : The data will be represented through a histogram to represent how the data is distributed

```python
sns.histplot(student_df['Exam_Score'], bins=15, kde=True)
plt.xlabel('Exam Scores')
plt.ylabel('Frequency')
plt.title('Distribution of Exam Scores')
plt.show()
```

**Overall Interpretation** : The graph is slightly skewed to the left, but the data is overall normalized. If we were to set a threshold for the passing to a score of 60, we could say that the students' general performance is good with the mean being 67.25, mode 68, and median of 67 these central tendency scores are above the threshold.

**2. Is there a difference in terms of the socioeconomic status of the students and are they correlated to one another?**

The socio-economic status will be determined through the access of wifi or internet at home and the family income level

First, this will be categorized on the internet access of the student then will generate the central tendency values of each category

```python
has_internet_access = student_df[student_df['Internet_Access']=='Yes']['Exam_Score']
no_internet_access = student_df[student_df['Internet_Access']=='No']['Exam_Score']

summary_data = {
    'Measure': ['Count', 'Mean', 'Median', 'Mode', 'Standard Deviation', 'Minimum', 'Maximum'],
    'Has Internet Access': [len(has_internet_access),
                            round(has_internet_access.mean(), 2),
                            round(has_internet_access.median(), 2),
                            round(has_internet_access.mode()[0], 2),
                            round(has_internet_access.std(), 2),
                            round(has_internet_access.min(), 2),
                            round(has_internet_access.max(),2)],
    'No Internet Access': [len(no_internet_access),
                           round(no_internet_access.mean(), 2),
                           round(no_internet_access.median(), 2),
                           round(no_internet_access.mode()[0], 2),
                           round(no_internet_access.std(),2),
                           round(no_internet_access.min(), 2),
                           round(no_internet_access.max(), 2)]
}

summary_table = pd.DataFrame(summary_data)
print('Performance of Students depending on their Internet Access')
display(summary_table)
```

Then we tried to see the differences on the students' performance based on the family incom

```python
low_income = student_df[student_df['Family_Income']=='Low']['Exam_Score']
medium_income = student_df[student_df['Family_Income']=='Medium']['Exam_Score']
high_income = student_df[student_df['Family_Income']=='High']['Exam_Score']

summary_data = {
    'Measure': ['Count', 'Mean', 'Median', 'Mode', 'Standard Deviation', 'Minimum', 'Maximum'],
    'Low Income': [len(low_income),
                            round(low_income.mean(), 2),
                            round(low_income.median(), 2),
                            round(low_income.mode()[0], 2),
                            round(low_income.std(), 2),
                            round(low_income.min(), 2),
                            round(low_income.max(),2)],
    'Medium Income': [len(medium_income),
                           round(medium_income.mean(), 2),
                           round(medium_income.median(), 2),
                           round(medium_income.mode()[0], 2),
                           round(medium_income.std(),2),
                           round(medium_income.min(), 2),
                           round(medium_income.max(), 2)],
    'High Income':[len(high_income),
                           round(high_income.mean(), 2),
                           round(high_income.median(), 2),
                           round(high_income.mode()[0], 2),
                           round(high_income.std(),2),
                           round(high_income.min(), 2),
                           round(high_income.max(), 2)]
}

summary_table = pd.DataFrame(summary_data)
print('Performance of Students depending on their Family Incom')
display(summary_table)
```

We will now be identifying the correlation of these socio-economic factors to the exam scores of the students

```python
student_df_corr = student_df.copy()

student_df_corr["Family_Income"] = (
    student_df_corr["Family_Income"].astype(str).str.strip().str.title().map({
        "Low": 1, "Medium": 2, "High": 3
    })
)

student_df_corr["Internet_Access"] = (
    student_df_corr["Internet_Access"].astype(str).str.strip().str.title().map({
        "No": 0, "Yes": 1
    })
)

corr=student_df_corr[['Exam_Score', 'Family_Income', 'Internet_Access']].corr(method='spearman')
print("Correlation Matrix")
display(corr)
```

The data above suggests that the family income and internet access or socio-economic factors have a weak correlation in with the students' academic performance. This leads us to change the focus of the study to other means that could correlate, determine, and predict the academic success of the students in the data set.

**4. Are study hours per week and access to educational materials correlated? And what are the differences in terms of the score from these factors?**

We will examine the relationship between study hours and exam scores through correlation analysis, and compare average exam scores across different levels of resource access.

```python
correlation = student_df['Hours_Studied'].corr(student_df['Exam_Score'])
print(f"Correlation between Study Hours and Exam Score: {correlation:.3f}")

resource_scores = student_df.groupby('Access_to_Resources')['Exam_Score'].mean()
print("\nAverage Exam Score by Resource Access:")
print(resource_scores)
```

```python
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.scatter(student_df['Hours_Studied'], student_df['Exam_Score'], alpha=0.5)
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.title('Study Hours vs Exam Score')

plt.subplot(1, 2, 2)
resource_scores.plot(kind='bar', color=['red', 'orange', 'green'])
plt.xlabel('Access to Resources')
plt.ylabel('Average Exam Score')
plt.title('Average Score by Resource Access')

plt.tight_layout()
plt.show()
```

There is a moderate positive correlation between study hours and exam scores. Students with better access to educational resources tend to achieve higher exam scores, showing that both study time and resource availability contribute to academic performance.

**5. How does socioeconomic status and study behaviors relate to student performance?**

We will analyze how family income and internet access relate to student performance by comparing average exam scores across these socioeconomic factors.

```python
income_scores = student_df.groupby('Family_Income')['Exam_Score'].mean()
print("Average Exam Score by Family Income:")
print(income_scores)

internet_scores = student_df.groupby('Internet_Access')['Exam_Score'].mean()
print("\nAverage Exam Score by Internet Access:")
print(internet_scores)
```

```python
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
income_scores.plot(kind='bar', color=['red', 'orange', 'green'])
plt.xlabel('Family Income')
plt.ylabel('Average Exam Score')
plt.title('Scores by Family Income')

plt.subplot(1, 2, 2)
internet_scores.plot(kind='bar', color=['red', 'green'])
plt.xlabel('Internet Access')
plt.ylabel('Average Exam Score')
plt.title('Scores by Internet Access')

plt.tight_layout()
plt.show()
```

Higher family income and internet access are associated with better exam scores, though the differences are relatively small. This suggests that while socioeconomic factors provide some advantage, individual study behaviors and effort may play a more significant role in determining academic success.
